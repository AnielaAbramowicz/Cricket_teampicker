```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
results = pd.read_csv('results.csv')
```

```python
# Calculate performance loss
results['loss'] = (results['Prior Performance (Actual)'] - results['Post Performance']) / results['Prior Performance (Actual)']
```

```python
# Plot histogram of loss
plt.figure(figsize=(10, 6))
sns.histplot(results['loss'], palette=sns.color_palette('muted', 1))
plt.xlabel('Performance Loss')
plt.ylabel('Frequency')
plt.show()
```
